```python
import pandas as pd

tabela = pd.read_csv("C:/Users/claud/Downloads/archive/all_indices_data.csv")
display(tabela)
```

```python
tabela = tabela[tabela ["volume"] >10000]
display(tabela)
```

```python
tabela["resultado"] = tabela ["close"] - tabela["open"]
display(tabela)
```

```python
tabela = tabela[tabela["resultado"] > 0]
display(tabela)
```

```python
tabela = tabela[tabela["open"] > 0]
display(tabela)
```

```python
tabela = tabela.sort_values(by=["ticker","resultado"], ascending=[True,False])
display(tabela)
```

```python
tabela = tabela.groupby('ticker')['resultado'].max()
display(tabela)
```

```python
tabela = tabela.sort_values(ascending=False)
display(tabela)
```
